```python
model = tf.keras.models.load_model("Mossbauer_model")
```

```python
espectro = -np.loadtxt("data.txt", delimiter=',')
espectro = espectro.T
espectro = (espectro - espectro.mean())/(espectro.std())
print(espectro.shape)
espectro = espectro.tolist()

print(espectro)
```

```python
espectro2 = -np.loadtxt('Hematita 15.txt', dtype=float)[:, 1]
print(espectro2.shape)
espectro2 = espectro2.tolist()


print(espectro2)
```

```python
categ = ["Otro", "Hematita", "Magnetita"]
```

```python
y2 = np.argmax(model.predict([espectro2]), axis=-1)

print(categ[y2[0]])
```
